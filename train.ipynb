```python
from datasets import load_dataset
import pandas as pd 

ds=load_dataset("electricsheepafrica/nigerian_realestate_commercial_properties",split="train")

df=pd.DataFrame(ds)


df.head()

```

```python
type(df['property_id'].tolist()[0])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

df['occupancy_status']=encoder.fit_transform(df['occupancy_status'])   # occupied =0 , vacant=1
df
```

```python
from sklearn.preprocessing import LabelEncoder

df['commercial_type'].value_counts()

encoder1=LabelEncoder()
df['commercial_type']=encoder1.fit_transform(df['commercial_type'])

df
```

```python
df['city'].unique()
```

```python
encoder1.classes_
```

```python
# feature engineering 
import numpy as np

df['rent_sqm']=(df['annual_rent']/df['size_sqm'])
df['lease_term_short']=np.where(df['lease_term_years']<2,1,0)
df['log_size_sqm']=np.log(df['size_sqm'])
```

```python
df.columns
```

```python
df=df[['property_id', 'listing_date', 'city', 'commercial_type', 'size_sqm', 'annual_rent',
       'occupancy_status', 'lease_term_years', 'rent_sqm',
       'lease_term_short', 'log_size_sqm']]
df.head()
```

```python
# engineering scoring features

df['city_median_rent_sqm']=df.groupby('city')['size_sqm'].transform('median')   # helps measure mispricing 
df['city_listing_count']=df.groupby('city')['property_id'].transform('count')   # buyer density / competition


```

```python
df.groupby('city')['property_id'].count()
```

```python
df.groupby('city')['size_sqm'].median()
```

```python
df.head()
```

```python
# mispricing score
df['mispricing_score']=abs(df['rent_sqm']-df['city_median_rent_sqm'])/df['city_median_rent_sqm']
df['mispricing_score']
```

```python
# df=df.drop(columns=['liquidity_score'])
# df

df['liquidity_score']=((1-(df['log_size_sqm']/df['log_size_sqm'].max())) + df['lease_term_short']+df['occupancy_status'])/3
df.head()
```

```python
df['velocity_score']=0.5*(df['liquidity_score']) + 0.5*(1/df['mispricing_score'])   # high score --> high velocity
df['ranking_score']=0.4*(df['velocity_score'])+0.3*(df['liquidity_score'])+0.3*(df['mispricing_score'])   # final score for ranking 
df=df.sort_values(by='ranking_score',ascending=False)
df.head(10)
```

```python
df.columns
```

```python
import xgboost as xgb  
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


cols=['rent_sqm','log_size_sqm','city_median_rent_sqm','city_listing_count','lease_term_short']   # don't use occupany status here ... (overfit)
X=df[cols]
Y=df['velocity_score']

velocity_model=XGBRegressor()
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=123)
velocity_model.fit(X_train,Y_train)

Y_pred=velocity_model.predict(X_test)

print(r2_score(Y_test,Y_pred))    
```

```python
import xgboost as xgb 
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

cols=['log_size_sqm','lease_term_short','city_listing_count','city_median_rent_sqm','commercial_type']
X=df[cols]
Y=df['ranking_score']

ranking_model=XGBRegressor()

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=123)

ranking_model.fit(X_train,Y_train)

Y_pred=ranking_model.predict(X_test)

r2_score(Y_test,Y_pred)
```

```python
import pandas as pd 

df['rank']=range(1,len(df)+1)


# neutral ranking for both buyers and sellers.

df['priority']=pd.qcut(       # 0 to 5% - immediate , 5 to 20% - High priority , 20% to 50%-Watchlist,50%-100% - Low priority
    df['rank'],
    q=[0,0.05,0.2,0.5,1.0],
    labels=['Immediate Action','High Priority','Watchlist','Low Priority']
)

df_front=df[['property_id', 'listing_date', 'city', 'commercial_type', 'size_sqm',
       'annual_rent', 'occupancy_status', 'lease_term_years', 'rent_sqm', 
        'priority']]


df_front['occupancy_status'].replace({1:'Vacant',0:'Occupied'},inplace=True)
df_front['commercial_type'].replace({0:'office',1:'retail_shop',2:'showroom',3:'warehouse'})


properties_json=df_front.to_dict(orient='records')

properties_json[:10]

```

```python
import json

file=open("properties.json","w")
json.dump(properties_json,file)
file.close()
```
